# Example for the Monte-Carlo simulation

* We use C++ code to generate the random numbers and estimate, if we node is inside the circle (red) or outside (blue)
* We use matplotlib to plot the results. This is not as easy in C++

```python
from py11 import py11, smap, svec
import matplotlib.pyplot as plt
```

```python
@py11(headers=["<random>"])
def comp()->svec[float]:
    """
    //Generate a random number device
    std::random_device rd;
    //Set the standard mersenne_twister_engine
    std::mt19937 gen(rd());
    std::uniform_real_distribution<double> dis(0,0.5);
    
    float x = dis(gen);
    float y = dis(gen);
    
    std::vector<float> res = {x,y,0};
    
    if ( x * x + y * y <= 0.25 )
        res[2] = 1;
    
    return res;
    """
```

```python
n=10
for i in range(0,n):
    res = comp();
    if res[2] == 0:
        plt.scatter(res[0],res[1],c="#5D8AA8")
    else:
        plt.scatter(res[0],res[1],c="#FE6F5E")

plt.xlabel("x")
plt.ylabel("y")
plt.xlim([0,0.5])
plt.title(str(n)+" random sample")
circle = plt.Circle((0., 0.), 0.5, color='gray',fill=False)
ax = plt.gca()
ax.add_artist(circle) 
```
